```python
import pandas as pd
import os
from IPython.display import display
```

```python
rents_data = "data/rents.csv"
rents_df = pd.read_csv(rents_data, sep=';', low_memory=False)
#pd.set_option('display.max_columns', None)
print(display(rents_df))
```

```python
# - .strip() removing leading/trailing whitespcaes
# - .replace() replace spaces with underscores
# - .lower() convert all uppercase to lowercase


rents_df.columns = rents_df.columns.str.lower().str.strip().str.replace(' ', '_')

#
rents_df.head()
```
